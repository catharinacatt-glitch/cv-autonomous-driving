import torch
import numpy as np

```python
data = [[1,2],[3,4]]
x_data = torch.tensor(data)
```

```python
###### tensor
```

```python
x_ones=torch.ones_like(x_data)
print(f"ones tensor: \n{x_ones} \n")
```

```python
x_rand=torch.rand_like(x_data, dtype=torch.float)
print(f"random tensor: \n {x_rand} \n")
```

```python
## tensor -> numpy
```

```python
t=torch.ones(5)
print(f"t: {t}")
```

```python
n=t.numpy()
print(f"n:{n}")
```

```python
t.add_(1)
print(f"{t}")
print(f"{n}")
```

```python
## numpy -> tensor
```

```python
n=np.ones(5)
t=torch.from_numpy(n)
```

```python
np.add(n,1,out=n)
print(f"t: {t}")
print(f"n: {n}")
```

```python
###### 연산 2
```

```python
a = torch.randn(3, 4)
b = torch.randn(3, 4)
print(f"a: {a}")
print(f"b: {b}")
```

```python
c = a + b            # 덧셈
d = a @ b.T          # 행렬곱 (matmul)
e = a * b            # 원소별 곱 
print(f"c: {c}\n")
print(f"d: {d}\n")
print(f"e: {e}")
```

```python
###### 연산 3
```

```python
x=torch.randn(2, 3, 4)
print(f"x:{x}\n")
y=x.view(6,4)
print(f"y:{y}\n")
z=x.permute(0,2,1)
print(f"z:{z}")
```
