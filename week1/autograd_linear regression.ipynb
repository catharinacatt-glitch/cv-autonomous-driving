```python
x=torch.tensor(3.0, requires_grad=True)  # x = 3.0
y=x ** 2 + 2*x +1     # y = x² + 2x + 1
y.backward()          # y'= 2x + 2
print(x.grad)         # y'= 2*3 + 2
```

```python
###### linear regression
```

```python
### y 예측모델
X=torch.randn(100, 1)
y_true=3 * X +1 +torch.randn(100, 1) * 0.1    # 정답
```

```python
### 파라미터 (y = wx + b)
w=torch.randn(1, requires_grad=True)
b=torch.zeros(1, requires_grad=True)
lr=0.1
```

```python
### 학습루프
for epoch in range(100):
    y_pred= X*w +b
    loss=((y_pred - y_true)**2).mean()  #MSE

    loss.backward()        #손실함수 미분

    with torch.no_grad():  #파라미터 업데이트
        w -= lr * w.grad
        b -= lr * b.grad
        
    w.grad.zero_()  # 초기화
    b.grad.zero_()

    if epoch % 10 ==0:
        print(f"epoch {epoch}: loss={loss.item():.4f}, w={w.item():.4f}, b={b.item():.4f}")
```
